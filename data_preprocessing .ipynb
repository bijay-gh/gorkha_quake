```python
from sklearn.preprocessing import OneHotEncoder, StandardScaler, MinMaxScaler
import pandas as pd

def preprocess_data(df):
    # Extract categorical features
    cat_cols = df.select_dtypes(include='object').columns
#     print(f"Categorical Feature:{cat_cols} /n length = {len(cat_cols)}")
    cat_df = df[cat_cols]
    
    # One-hot encode categorical features
    encoder = OneHotEncoder(handle_unknown='ignore')
    cat_df_encoded = encoder.fit_transform(cat_df)
    cat_df_encoded = pd.DataFrame(cat_df_encoded.toarray(), columns=encoder.get_feature_names(cat_cols), index = df.index)
    
    # Extract numerical features
    num_cols = df.select_dtypes(include=['int64']).columns
    num_df = df[num_cols]
    
    # Identify and normalize binary features
    binary_cols = [col for col in num_df.columns if num_df[col].nunique() == 2]
#     print(f"Binary Feature:{binary_cols} /n length = {len(binary_cols)} ")
    if binary_cols:
        for col in binary_cols:
            num_df[col] = num_df[col].astype(bool).astype(int)
            scaler = MinMaxScaler()
            num_df[col] = scaler.fit_transform(num_df[col].values.reshape(-1, 1))
    
    num_col = num_df.drop(binary_cols, axis=1).columns
#     print(f"Numerical Feature:{num_col}/n length = {len(num_col)} ")
    # Scale numerical features
    scaler = StandardScaler()
    num_df_scaled = scaler.fit_transform(num_df.drop(binary_cols, axis=1))
    num_df_scaled = pd.DataFrame(num_df_scaled, columns=num_cols.drop(binary_cols),index = df.index)
    
    # Combine encoded categorical, binary and scaled numerical features into a new dataframe
    df_preprocessed = pd.concat([num_df_scaled, cat_df_encoded, num_df[binary_cols]], axis=1)
    
    return df_preprocessed

```

```python
X = train.drop('damage_grade',axis=1)
y = train['damage_grade']
X_scaled = preprocess_data(X)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.25, random_state=5)
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)
rf_classifier.fit(X_train, y_train)
```

#  Gradient Boosting Classifier without winsorizing

```python
from sklearn.ensemble import GradientBoostingClassifier
gb_classifier = GradientBoostingClassifier(n_estimators=800, learning_rate=0.2, max_depth=9,min_samples_split=1600,min_samples_leaf=250,subsample=0.9)

# Fit the model to the training data
gb_classifier.fit(X_train, y_train)
```

```python
# Make predictions on the test data
y_pred = gb_classifier.predict(X_test)
```

```python
# Calculate the F1 score micro
f1_micro = f1_score(y_test, y_pred, average='micro')
print(f"F1 Score Micro: {f1_micro}")
```

```python
test_scaled = preprocess_data(test)
prediction = gb_classifier.predict(test_scaled)
```

```python
# Ensure the predictions are in teh correct submission format
submission_format = pd.read_csv('submission_format.csv',index_col = 'building_id')
result = pd.DataFrame(data=prediction,
                             columns=submission_format.columns,
                             index=submission_format.index)
result['damage_grade'] =result['damage_grade'].astype(int)
result.head()
```
